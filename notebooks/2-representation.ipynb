```python
from datasets import load_from_disk

# Load the preprocessed dataset
new_ds = load_from_disk("/content/drive/MyDrive/Lighthouse Labs/LLM-Project/notebooks")
```

```python
from transformers import AutoTokenizer

# Using DistilBERT for sentiment analysis
model_checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
# Tokenization function for classification task
def tokenize_function(example):
    return tokenizer(
        example["text"],
        padding="max_length",
        truncation=True,
        max_length=512
    )
```

```python
wha# Apply tokenization
tokenized_dataset = new_ds.map(tokenize_function, batched=True)
```

```python
# Check a sample
print(tokenized_dataset["train"][0])
```

```python
# Save the tokenized dataset for training
tokenized_dataset.save_to_disk("/content/drive/MyDrive/Lighthouse Labs/LLM-Project/notebooks/tokenized_imdb")
```
